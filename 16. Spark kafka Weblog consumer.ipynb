```python
import sys
import os
import sqlite3
from pyspark import SparkContext, SparkConf
from pyspark.streaming import StreamingContext
from pyspark.streaming.kafka import KafkaUtils
```

```python
os.environ['PYSPARK_SUBMIT_ARGS'] = '--packages org.apache.spark:spark-streaming-kafka-0-8_2.11:2.3.2 pyspark-shell'
```

```python
sc = SparkContext(appName='streamTest')
ssc = StreamingContext(sc, 5)
```

```python
def insertData( data ):
    sql = "insert into ktest values(?,?)"
    db = sqlite3.connect(r'C:\my.db')
    db.execute( sql, data  )
    db.commit()
    db.close()
```

```python
#topic 이 있으면 호출
def recvData( data):
    insertData( data )
    print("data=",data) #('키','값')
    return data
```

```python
topic = 'sample'
kvs = KafkaUtils.createStream(ssc, 'localhost',
    'raw-event-streaming-consumer', {topic:1} )
#========================
#메시지 수신시 recvData call
lines = kvs.map( recvData ) #콜백함수등록
lines.pprint() #상태메시지...
#==============================
ssc.start() #폴링 시작( 5초에 한번씩 카프카클러스터에 sample topic 확인)
ssc.awaitTerminationOrTimeout(2)
```

```python
ssc.stop() #폴링중지
```
